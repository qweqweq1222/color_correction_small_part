```python
import numpy as np
import matplotlib.pyplot as plt
Pts_src =[[-0.0524562,-0.0147996],
[-0.0642142,0.0112036],
[-0.427621,0.444681],
[-0.189359,-0.251346],
[-0.178773,-0.281382]
]
Pts_dst =[[-0.0504605,-0.0131448],
[-0.0594375,0.00789988],
[-0.434893,0.414712],
[-0.477951,0.237258],
[-0.27392,-0.349971],
[-0.168219,-0.263366]
]

Pts_src = np.array(Pts_src)
Pts_dst = np.array(Pts_dst)

X_src = Pts_src[:,0]
Y_src = Pts_src[:,1]
X_dst = Pts_dst[:,0]
Y_dst = Pts_dst[:,1]

fig = plt.figure()
fig.set_size_inches(10,10)
plt.plot(X_dst,Y_dst, label = 'dst')
plt.plot(X_src,Y_src, label = 'src')
plt.legend()
```

```python
Pts_final = []
matrix = np.array([[1.46535,-0.0338861],[-0.000851406,1.01245]])
for p in Pts_src:
    Pts_final.append(matrix.dot(p)  + np.array([0.0378197,-0.0334211]))
Pts_final = np.array(Pts_final)
X_final = Pts_final[:,0]
Y_final = Pts_final[:,1]
fig = plt.figure()
fig.set_size_inches(10,10)
plt.plot(X_src,Y_src, label = 'src')
plt.plot(X_dst,Y_dst, label = 'dst')
plt.plot(X_final,Y_final, label = 'res')
plt.legend()
```

```python
with open("covar23.csv") as file:
    lines = file.readlines()
B_ = []
for line in lines:
    B_.append(float(line))
fig = plt.figure()
fig.set_size_inches(10,10)
plt.hist(B_, density=False, bins=30, range = [0, 0.06])
plt.ylabel('amount')
plt.xlabel('covar')
plt.grid()
```
